**Анализ текстов**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# статья 1 и 4 - о сноуборде
# статья 2 и 3 - о хокее
#
#

c_names = ['post1', 'post2', 'post3', 'post4']
words = ['ice', 'snow', 'hockey', 'goal', 'puck']
post_words = pd.DataFrame([[4, 4, 6, 2],
                           [6, 1, 0, 5],
                           [3, 0, 0, 5],
                           [0, 6, 5, 1],
                           [0, 4, 5, 0]],
                          index = words,
                          columns = c_names)
post_words.index.names = ['word:']
post_words
```

```python
U, sigma, V = np.linalg.svd(post_words)
print ("V = ")
print (np.round(V, decimals=2))
```

```python
# Видно, что посты 1,4 и 2,№ похожи между собой
V_df = pd.DataFrame(V, columns=c_names)   # выведем собственные векторы
V_df
```

```python
# Видно, что первые сигма несут основную информацию, остальные несут незначительную информацию, которую можно отбросить
sigma
```

```python
n_approx = 2
A_approx = np.matrix(U[:, :n_approx]) * np.diag(sigma[:n_approx]) * np.matrix(V[:n_approx, :])

print ("A calculated using truncated matrix S:\n")
print (pd.DataFrame(A_approx, index=words, columns=c_names))
print ("\nError from actual value:\n")
print (post_words - A_approx)
```

```python
# Для наглядности визуализируем числа и посмотрим на строки
%matplotlib inline
import matplotlib.pyplot as plt

plt.xticks(range(len(c_names)))
plt.yticks(range(len(words)))
plt.ylim([len(words) - 1.5, -.5])
ax = plt.gca()
ax.set_xticklabels(c_names)
ax.set_yticklabels(range(1, len(words) + 1))
plt.title("$V$")
plt.imshow(V, cmap='gray');
plt.colorbar();
```

```python
# Видно, насколько разные темы в зависимости от количества вхождений различных слов
pd.DataFrame(U[:,1], index=words)
```

**Получение ковариационной матрицы**

```python
import numpy as np
from scipy.linalg import eigh, cholesky
from scipy.stats import norm

from pylab import plot, show, axis, subplot, xlabel, ylabel, grid


# Choice of cholesky or eigenvector method.
method = 'cholesky'
#method = 'eigenvectors'

num_samples = 1000

# The desired covariance matrix.
r = np.array([
        [ 1,    0.5,    0],
        [ 0.5,  1,      0],
        [ 0,    0,      2]
    ])

# Generate samples from three independent normally distributed random
# variables (with mean 0 and std. dev. 1).
x = norm.rvs(size=(3, num_samples))

    # Compute the Cholesky decomposition.
c = np.linalg.cholesky(r)

# Convert the data to correlated random variables. 
y = np.dot(c, x)

#
# Plot various projections of the samples.
#
subplot(2,2,1)
plot(y[0], y[1], 'b.')
ylabel('y[1]')
axis('equal')
grid(True)

subplot(2,2,3)
plot(y[0], y[2], 'b.')
xlabel('y[0]')
ylabel('y[2]')
axis('equal')
grid(True)

subplot(2,2,4)
plot(y[1], y[2], 'b.')
xlabel('y[1]')
axis('equal')
grid(True)

show()
```

```python
U, S, V = np.linalg.svd(y)   # Получаем ковариацию через SVD
covSVD = U.dot(np.diag(S**2).dot(U.T))/num_samples
print('Ковариация:\n',covSVD)
print('\nСобств. значения:\n',(S**2)/num_samples)
print('\nСобств. векторы:\n',U)
```

```python
covC = y.dot(y.T)/num_samples      # Получаем ковариацию в лоб
eigC = np.linalg.eig(covC)
print('Ковариация:\n',covC)
print('\nСобств. значения:\n',eigC[0])
print('\nСобств. векторы:\n',eigC[1])
```
